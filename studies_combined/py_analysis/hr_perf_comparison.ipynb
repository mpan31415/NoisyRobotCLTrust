```python
from pandas import read_csv
from os import getcwd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

STUDY1_RESULTS_PATH = getcwd() + "\\..\\..\\study1_results\\"
STUDY2_RESULTS_PATH = getcwd() + "\\..\\..\\study2_results\\"
```

```python
df1 = read_csv(STUDY1_RESULTS_PATH + "all_data.csv", index_col=False)
df2 = read_csv(STUDY2_RESULTS_PATH + "all_data.csv", index_col=False)
ti1 = read_csv(STUDY1_RESULTS_PATH + "traj_csv_logs\\tap_impact.csv", index_col=False)
```

#### Compare the performance between {human-only} and {robot-only}. 
- We should expect to see similar performance, as the noisy robot should have a similar error magnitude as the human on average.

```python
####### STUDY 1 RESULTS #######

no_auto_list = df1[df1['autonomy']==0.0]['overall_traj_err'].tolist()
full_auto_list = df1[df1['autonomy']==0.0]['robot_traj_err'].tolist()

e1 = sum(no_auto_list) / len(no_auto_list)
e2 = sum(full_auto_list) / len(full_auto_list)

print("###### RESULTS FROM STUDY 1: ######")
print("Average human-only error: %.3f" % e1)
print("Average robot-only error: %.3f" % e2)

plt.boxplot([no_auto_list, full_auto_list])
plt.show()
```

```python
####### STUDY 2 RESULTS #######
no_auto_list = df2[df2['autonomy']==0.0]['overall_traj_err'].tolist()
full_auto_list = df2[df2['autonomy']==0.0]['robot_traj_err'].tolist()

e1 = sum(no_auto_list) / len(no_auto_list)
e2 = sum(full_auto_list) / len(full_auto_list)

print("###### RESULTS FROM STUDY 2: ######")
print("Average human-only error: %.3f" % e1)
print("Average robot-only error: %.3f" % e2)

plt.boxplot([no_auto_list, full_auto_list])
plt.show()
```

```python
exp1_human_list = df1[df1['autonomy']==0.0]['overall_traj_err'].tolist()
exp2_robot_list = df2[df2['autonomy']==0.0]['robot_traj_err'].tolist()

e1 = sum(exp1_human_list) / len(exp1_human_list)
e2 = sum(exp2_robot_list) / len(exp2_robot_list)

res = f_oneway(exp1_human_list, exp2_robot_list)
print(res)
print("")

print("###### COMPARISON BETWEEN STUDIES 1 & 2: ######")
print("Average human-only error in study 1: %.3f" % e1)
print("Average robot-only error in study 2: %.3f" % e2)

plt.boxplot([exp1_human_list, exp2_robot_list])
plt.title("Error Comparison between exp1 human baseline and exp2 robot")
plt.show()
```

```python
exp1_human_list = ti1['base_err'].tolist()
exp2_robot_list = df2[df2['autonomy']==0.0]['robot_traj_err'].tolist()

e1 = sum(exp1_human_list) / len(exp1_human_list)
e2 = sum(exp2_robot_list) / len(exp2_robot_list)

res = f_oneway(exp1_human_list, exp2_robot_list)
print(res)
print("")

print("###### COMPARISON BETWEEN STUDIES 1 & 2: ######")
print("Average human-only error in study 1: %.3f" % e1)
print("Average robot-only error in study 2: %.3f" % e2)

plt.boxplot([exp1_human_list, exp2_robot_list])
plt.title("Error Comparison between exp1 human baseline and exp2 robot")
plt.show()
```
